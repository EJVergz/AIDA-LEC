```python
 def plotHorzBar(column):
  data = flood_df[column]
  count = {}
  for d in data:
    words = d.split(";")
    for w in words:
      if w in count.keys():
        count[w] = count[w] + 1
      else:
        count[w] = 1
  
  y_pos = np.arange(len(count.keys()))
  plt.barh(y_pos,count.values())
  plt.yticks(y_pos, count.keys())
  plt.show()
```

```python
plotHorzBar("Why should you study about flooding caused by structural floods?")
```

```python
plotHorzBar("What may be causing the floods in your area?")
```

```python
plotHorzBar("What do you think the is most effective solution?")
```
